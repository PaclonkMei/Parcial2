```python
dfc = pd.read_csv("clientes.csv")
dfo = pd.read_csv("ordenes.csv")
```

```python
df = pd.read_csv("sensores.csv", usecols=['ts', 'temperatura', 'humedad'])
```

```python
df['ts'] = pd.to_datetime(df['ts'], format='mixed', errors='coerce')
```

```python
mediana_temp = df['temperatura'].median()
media_hum = df['humedad'].mean()
```

```python
df['temperatura'] = df['temperatura'].fillna(mediana_temp)
df['humedad'] = df['humedad'].fillna(media_hum)
```

```python
print("NaT en ts:", df['ts'].isna().sum())
print("Mediana final temperatura:", round(df['temperatura'].median(), 1))
print("Media final humedad:", round(df['humedad'].mean(), 3))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
textos = [
   "el gato duerme",
   "el perro ladra",
   "el gato maulla",
   "el perro corre"
]

y = ["gato", "perro", "gato", "perro"]
```

```python
modelo = Pipeline([
   ("vect", CountVectorizer()),
   ("clf", MultinomialNB())
])
```

```python
modelo.fit(textos, y)
```

```python
pred = modelo.predict(["el gato corre"])
print(pred)
```
